```python
def disparity(im1, im2, bks = 10 , sr = 50, ncc= True):
    
    hbs = int(np.floor(bks/2))
    disp = np.zeros(im1.shape)
    
    for r1 in range(hbs, im1.shape[0] -hbs):
        if r1%10 == 0:
            print(r1)
        for c1 in range(hbs, im1.shape[1] - hbs):
            
            if ncc:
                DIST = -np.inf
                
            else:
                DIST = np.inf

            
            b1 = im1[r1 - hbs:r1+hbs + 1, c1-hbs: c1+hbs+1]

            #look for the block
            start_index = np.max((hbs, c1-sr))
            stop_index = np.min((im1.shape[1]-hbs, c1+sr))
           
            for c2 in range(start_index, stop_index):

                b2 = im2[r1-hbs: r1+hbs+1, c2 - hbs: c2 +hbs+1 ]
                
                
                #subtract the mean from each block
                b1m = b1.ravel() - b1.mean()
                b2m = b2.ravel() - b2.mean()
                
                #ncc
                if ncc:
                    value = b1m.dot(b2m) / np.sqrt( (b1m**2).dot(b2m**2))
                    if value > DIST:
                        DIST = value
                        #best_box = b2
                        best_c = c2
                        
                #ssd
                else:
                    value = ((b1 - b2)**2).sum()
                    if value < DIST:
                        DIST = value
                        #best_box = b2
                        best_c = c2
                               
            disp[r1,c1]=np.abs(c1-best_c)
                
     
                            
#             print('distance', value)
#             plt.imshow(b1, cmap = 'gray')
#             plt.title("original ({},{})".format(r1,c1))
#             plt.axis('off')
#             plt.show()
#             plt.imshow(best_box, cmap = 'gray')
#             plt.title("match ({},{})".format(r1,best_c))
#             plt.axis('off')
#             plt.show()


            
    return disp
                
    
```

```python
test = disparity(I2, I6 )
```
